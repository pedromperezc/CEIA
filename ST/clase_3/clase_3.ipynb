# DATASET 1 TRANSACCIONES REALIZADAS:

Este dataset representa el importe recaudado por diferentes empresas a través de un botón de pago.


```python
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d')
```

```python
df = pd.read_csv("/content/drive/MyDrive/TIME_SERIES/pagos.csv",parse_dates=['fecha'], date_parser=dateparse)
```

```python
sns.lineplot(data=df[(df.agente == 3349)&(df.fecha > '2021-01-01')&(df.fecha < '2021-12-31')], x="fecha", y="importe", hue='agente', palette= 'deep')
```

```python
sns.lineplot(data=df[(df.agente != 854) &(df.fecha > '2021-04-01')&(df.fecha < '2021-05-30')], x="fecha", y="importe", hue='agente', palette= 'deep')
```

**Observaciones**:

* Se observa un leve incremento en los picos.

* Se observa un comportamiento de estacionalidad reflejado en un aumento en el total procesado los primeros días del mes, y una disminución los fines de semana.
